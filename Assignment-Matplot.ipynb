```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
AirFares = pd.read_csv(r'C:\Users\Public\Narayanan\Airfares.csv')
AirFares.head(5)
```

```python
plt.figure(figsize=(15,18))
plt.title('Bar Chart showing people travelling for vacation ')
sns.barplot(y=AirFares['Arrival_CITY'], x=AirFares['Arrival_avg_population'], hue = AirFares['VACATION'])
```

```python
#Story: The people travel for their vacation and do not travel for vacation seems to be same in almost all the cities except
# Chicago, Columbus, Tampa, Durham.  Big cities like Los Angels, New York seems to be exactly equal in both travelling and not
# travelling
```

```python
plt.figure(figsize=(15,18))
plt.title('Box Plot showing the people preferences in the cities they travel and Kms they travel for that ')
sns.boxplot(x=AirFares["DISTANCE"],y=AirFares['Arrival_CITY'])
```

```python
# Story:  This chart tells about people travelling to major cities.  It also shows that people from major cities found to be
# travelling to other major cities.  No one likes to travel to small cities in their vacation.  Though those cities are costly,
# still they prefer that 
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(AirFares.corr(), xticklabels=AirFares.corr().columns, yticklabels=AirFares.corr().columns, 
            cmap='RdYlGn', center=0, annot=True)
```

```python
# Story:  The above pictures show the correlation between all possible pairs.  Some are negatively correlated and some are 
# positively correlated.
```

```python
Counts = AirFares.groupby(['Departure_avg_INCOME']).size().reset_index(name='counts')
fig, ax = plt.subplots(figsize=(10,10), dpi= 80)    
sns.stripplot(Counts.Departure_avg_INCOME, size=Counts.counts*2, ax=ax)
```

```python
# Story:  This diagram show the concentration of people in the 25000 and 35000 group has the most Departure
```
